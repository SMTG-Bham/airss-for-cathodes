## NEB result plot

```python
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path
plt.rcParams["figure.dpi"] = 100
```

```python
def plot_data(workdir, color='medium blue'):
    data = np.loadtxt(str(Path(workdir) / 'spline.dat'))
    #idx, coord, eng, tang, _ = data.T
    idx, coord, eng, tang = data.T
    plt.figure(dpi=100, figsize=(3,3))
    plt.plot(coord, eng, '-', color=f'xkcd:{color}')

    data =  np.loadtxt(str(workdir / 'neb.dat'))
    idx, coord, eng, tang, _ = data.T
    plt.plot(coord, eng, 's', color=f'xkcd:{color}')
    plt.ylim(0, 0.5)
    plt.xlabel('Reaction coordinates')
    plt.ylabel('Energy barrier (eV)')
    plt.tight_layout()
    return plt
```

```python
plot_data(Path("fm-elec-plus-one/vac36_path1-climb"))
#plt.savefig("neb-path1-nelect.svg")
```

```python
plot_data(Path("fm-elec-plus-one/vac36_path2"), 'red')
#plt.savefig("neb-path2-nelect.svg")
```
